```python
import geopandas as gpd
```

```python
karnataka_districts = gpd.read_file('/content/drive/MyDrive/KSP/District.shp')

fig, ax = plt.subplots(figsize=(10, 10))
karnataka_districts.plot(ax=ax, edgecolor='black', linewidth=0.5)
ax.set_title('Map of Karnataka')
plt.axis('off')
plt.show()
```

```python
print(karnataka_districts)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
karnataka_districts.plot(ax=ax, edgecolor='black', linewidth=0.5)

for idx, row in karnataka_districts.iterrows():
    district_name = row['KGISDist_1']
    centroid_x, centroid_y = row['geometry'].centroid.x, row['geometry'].centroid.y
    ax.text(centroid_x, centroid_y, district_name, fontsize=8, ha='center', va='center')

ax.set_title('Map of Karnataka with District Names')
plt.axis('off')
plt.show()
```

```python
geo_district_names = set(karnataka_districts['KGISDist_1'])
dataset_district_names = dataset['District_Name'].unique().tolist()
```

```python
matching_districts = set(dataset_district_names).intersection(geo_district_names)
mismatching_districts = set(dataset_district_names).difference(geo_district_names)

print("Matching districts found in GeoDataFrame:")
print(matching_districts)
print("\nDistricts in dataset not found in GeoDataFrame:")
print(mismatching_districts)
```

```python
district1_counts = dataset[dataset['District_Name'].isin(matching_districts)]['District_Name'].value_counts()
district1_counts
```

```python
matched_district_names = district1_counts.index.tolist()
matched_district_counts = district1_counts.values.tolist()
```

```python
dataset_district_victims = {
    'District_Name1': matched_district_names,
    'Victims_Count1': matched_district_counts
}

# Convert the dictionary to a DataFrame
district_counts_df = pd.DataFrame(dataset_district_victims)

# Merge the counts with the GeoDataFrame
karnataka_districts = karnataka_districts.merge(district_counts_df, left_on='KGISDist_1', right_on='District_Name1', how='left')
karnataka_districts = karnataka_districts.fillna(0)

# Plot the chloropleth map
fig, ax = plt.subplots(figsize=(10, 10))
karnataka_districts.plot(column='Victims_Count1', cmap='Blues', ax=ax, edgecolor='black', linewidth=0.5, legend=True)

# Add district names to the map
for idx, row in karnataka_districts.iterrows():
    x, y = row.geometry.centroid.coords.xy
    ax.annotate(row['KGISDist_1'], xy=(x[0], y[0]), fontsize=8, ha='center')

# Add names of districts not merged
not_merged_districts = set(karnataka_districts['KGISDist_1']) - set(matched_district_names)
for district in not_merged_districts:
    district_geometry = karnataka_districts[karnataka_districts['KGISDist_1'] == district].geometry.values[0]
    x, y = district_geometry.centroid.coords.xy
    ax.annotate(district, xy=(x[0], y[0]), fontsize=8, ha='center', color='red')

ax.axis('off')
ax.set_title('District Counts', fontsize=16)
plt.show()
```

```python
age_bins = [0, 20, 40, 60, 80, 120]
labels = ['0-20', '21-40', '41-60', '61-80', '80+']
age_groups = pd.cut(dataset['age'], bins=age_bins, labels=labels, include_lowest=True, right=False)

age_group_counts = age_groups.value_counts().reset_index()

age_group_counts.columns = ['age_group', 'count']

fig, ax = plt.subplots(figsize=(6, 7))
ax.pie(age_group_counts['count'], labels=age_group_counts['age_group'], autopct='%1.1f%%', startangle=90)
ax.axis('equal')
ax.set_title('Age Group Distribution', fontsize=16)
plt.show()
```
